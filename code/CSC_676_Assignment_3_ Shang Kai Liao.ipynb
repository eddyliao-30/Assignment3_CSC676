```
#Q1
x=np.linspace(0.0, 50)# Create samples from 0 to 50
y=325*np.sin(2*np.pi*50*x)# Calculation sample value by the equation

plt.plot(x,y)# plot signal out
discrete_data=np.sin(2 * np.pi * 50 * x )*325# Calculate the discrete data
plt.stem(x, discrete_data)# Create stem plot of the sigal
plt.show()
```

```
#Q2
t=np.linspace(-0.02,0.05,100) # setting 100 samples in -0.02 to 0.05
x=np.exp(1j*100*np.pi*t) # calculate the eqation

plt.title('real part')
plt.plot(t,x.real) # plot the real part of the eqation
plt.show()

plt.title('imaginary part')
plt.plot(t,x.imag) # plot the image part of the eqation
plt.show()
```

```
#Q3 original single
N=32 # setting N
t=np.linspace(0,20,N) # set up sample list
x=np.exp(-t/3)*np.cos(2*t) #calculate the equation
plt.stem(t,x) # plot the stem figure out
plt.show()



f=np.fft.fft(x) # do the 1d fft
plt.stem(t,f.real,'r',markerfmt='ro') # plot the real part of fft result
plt.stem(t,f.imag,'g',markerfmt='go') # plot the imag part of fft result
plt.show()
plt.stem(t,abs(f)) # plot the magnitude of fft result
plt.show()
```

```
#Q5
import numpy as np
import matplotlib.pyplot as plt

N,M=20,20
A=255//2
u,v=2,0
def sine_wave_calculation(u,v):
  img=np.zeros((N,M)) # set up a N*M size matrix
  for m in range(M):
    for n in range(N):
      img[m,n]=A*np.sin(2*np.pi*(u*n/N+v*m/M)) # do the equation A*sin(2*pi*(u*n/N+v*m/M))
  return img[::-1] # inverse column 


plt.subplot(121)
plt.imshow(sine_wave_calculation(2,0),cmap='gray')
plt.subplot(122)
magnitude=np.fft.fftshift(np.fft.fft2(sine_wave_calculation(2,0)))
plt.imshow(abs(magnitude),cmap='gray')
plt.show()

plt.subplot(121)
magnitude=np.fft.fftshift(np.fft.fft2(sine_wave_calculation(3,1)))
plt.imshow(sine_wave_calculation(3,1),cmap='gray')
plt.subplot(122)
plt.imshow(abs(magnitude),cmap='gray')
plt.show()

plt.subplot(121)
magnitude=np.fft.fftshift(np.fft.fft2(sine_wave_calculation(7,-5)))
plt.imshow(sine_wave_calculation(7,-5),cmap='gray')

plt.subplot(122)
plt.imshow(abs(magnitude),cmap='gray')
plt.show()
plt.tight_layout()
```

```
#6 the codes provided by professor
import numpy as np
import scipy.signal as sig
from scipy import misc
import matplotlib.pyplot as plt
from scipy import ndimage
import imageio

kernel = (1.0/256)*np.array([[1, 4,  6,  4,  1],[4, 16, 24, 16, 4],[6, 24, 36, 24, 6],[4, 16, 24, 16, 4],[1, 4,  6,  4,  1]])

def interpolate(image):
    """
    Interpolates an image with upsampling rate r=2.
    """
    image_up = np.zeros((2*image.shape[0], 2*image.shape[1]))
    # Upsample
    image_up[::2, ::2] = image
    # Blur (we need to scale this up since the kernel has unit area)
    # (The length and width are both doubled, so the area is quadrupled)
    #return sig.convolve2d(image_up, 4*kernel, 'same')
    return ndimage.filters.convolve(image_up,4*kernel, mode='constant')
                                
def decimate(image):
    """
    Decimates at image with downsampling rate r=2.
    """
    # Blur
    #image_blur = sig.convolve2d(image, kernel, 'same')
    image_blur = ndimage.filters.convolve(image,kernel, mode='constant')
    # Downsample
    return image_blur[::2, ::2]                                
               
                                                 
# here is the constructions of pyramids
def pyramids(image):
    """
    Constructs Gaussian and Laplacian pyramids.
    Parameters :
        image  : the original image (i.e. base of the pyramid)
    Returns :
        G   : the Gaussian pyramid
        L   : the Laplacian pyramid
    """
    # Initialize pyramids
    G = [image, ]
    L = []

    # Build the Gaussian pyramid to maximum depth
    while image.shape[0] >= 2 and image.shape[1] >= 2:
        image = decimate(image)
        G.append(image)
        
    # Build the Laplacian pyramid
    for i in range(len(G) - 1):
        L.append(G[i] - interpolate(G[i + 1]))

    return G[:-1], L
```

```
#6 
def bond(a,b): # Bond two image and limit the values will not over 255
  r=a+b
  for i in range(len(r)):
    for j in range(len(r[0])):
        if r[i,j]>=255:
          r[i,j]=255
  return r

def blending(G1last,L1,G2last,L2,Gm,Lm): # to blend the two images by laplacian mathod 
  
  L=len(Lm)
  G1_next=G1last.copy()
  G2_next=G2last.copy()
  mask=Gm.copy()
  for j in range(L-2,-1,-1):
    mask=interpolate(mask+Lm[j+1])
    G1_next=interpolate(G1_next)+L1[j]*mask
    G2_next=interpolate(G2_next)+L2[j]*(1-mask)
  
  return bond(G1_next, G2_next)

```

```
def main(img1,img2,mask): # the main function for laplacian mathod 
  # extra image's r,g,b and create its Gausse and Laplacian pyramids
  [G_1r,L_1r] = pyramids(img1[:,:,0])
  [G_1g,L_1g] = pyramids(img1[:,:,1])
  [G_1b,L_1b] = pyramids(img1[:,:,2])

  [G_2r,L_2r] = pyramids(img2[:,:,0])
  [G_2g,L_2g] = pyramids(img2[:,:,1])
  [G_2b,L_2b] = pyramids(img2[:,:,2])

  [G_m,L_m] = pyramids(mask/np.max(mask))

  reconstruct_r=blending(G_1r[-1],L_1r,G_2r[-1],L_2r,G_m[-1],L_m) # processing blending
  reconstruct_g=blending(G_1g[-1],L_1g,G_2g[-1],L_2g,G_m[-1],L_m)
  reconstruct_b=blending(G_1b[-1],L_1b,G_2b[-1],L_2b,G_m[-1],L_m)
  img=img1.copy() # create a 512,512,3 matrix as container
  img[:,:,0],img[:,:,1],img[:,:,2] = reconstruct_r,reconstruct_g,reconstruct_b # store the value into the container
  return img
```

```
apple=imageio.imread('apple.jpg')
orange=imageio.imread('orange.jpg')
mask=imageio.imread('mask.jpg',as_gray=True)


img=main(apple,orange,mask)
plt.imshow(img)
imageio.imwrite("apple orange blended.jpg",img)
```
